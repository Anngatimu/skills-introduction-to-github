```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("Desktop\exams data.csv")
data
```

```python
data.describe()
```

```python
data.dropna()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(data['math score'], data['reading score'])
plt.title(' Line Plot: Math Score vs Reading Score')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```

```python
condition = data['test preparation course'] == 'completed'
plt.figure(figsize=(10, 10))
plt.plot(data[condition]['math score'], data[condition]['reading score'], color='blue', label='Completed')
plt.plot(data[~condition]['math score'], data[~condition]['reading score'], color='red', label='Not Completed')

plt.title('Line Plot: Math Score vs Reading Score')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(x='math score', y='reading score', hue='test preparation course',data=data,palette='Set2')
plt.title('Scatter Plot: Math Score vs Reading Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['math score'], bins=30, edgecolor='red')
plt.title('Histogram of Math Scores')
plt.xlabel('Math Score')  # Labeling the x-axis
plt.ylabel('Frequency')  # Labeling the y-axis
plt.grid(True)  # Adding grid lines for better readability
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['math score'], bins=30, edgecolor='red')
plt.title('Histogram of Math Scores')
plt.xlabel('Math Score')  # Labeling the x-axis
plt.ylabel('Frequency')  # Labeling the y-axis
plt.grid(True)  # Adding grid lines for better readability
plt.show()
Q1 = data['math score'].quantile(0.25)
Q3 = data['math score'].quantile(0.75)
IQR = Q3 - Q1
outliers = data[(data['math score'] < Q1 - 1.5 * IQR) | (data['math score'] > Q3 + 1.5 * IQR)]
outliers
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['math score'])
plt.title('Boxplot of Math Scores')
plt.xlabel('Math Score')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Desktop\cleaned_sbt_japan.csv")
df.head(16)
```

```python
#what are the top 10 manufacturers with themost cars
top_manufacturers=df["car_brand"].unique()
print(top_manufacturers)
```

```python
top_manufacturers=df["car_brand"].value_counts()
print(top_manufacturers)
```

```python
top_manufacturers=df["car_brand"].value_counts().max()
print(top_manufacturers)
```

```python
top_manufacturers=df["car_brand"].value_counts().nlargest(10)
print(top_manufacturers)
```

```python
#begin the visualizationusing std fig size of(10,8)
plt.figure(figsize=(20,10))
sns.barplot(x=top_manufacturers.index,y=top_manufacturers.values)
plt.show()
```

```python
top_manufacturers = df['car_brand'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_manufacturers.index, y=top_manufacturers.values,palette='viridis' )
plt.title('TOP MANUFACTURERS IN JAPAN')
plt.xlabel('car_brand')
plt.ylabel('no.of cars manufactured')
plt.show()
```

```python
#graph showing the top 5 Toyota models
top_toyota_models=df[df['car_brand']=='TOYOTA']['car_model'].value_counts().nlargest(5)
print(top_toyota_models)
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=top_toyota_models.index,y= top_toyota_models.values,palette='viridis')
plt.title('top_toyota_models')
plt.xlabel('toyota brands')
plt.ylabel('no. of toyota manufactured')
plt.show()
```
